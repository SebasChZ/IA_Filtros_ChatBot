```python
from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt

imgCovid = cv2.imread("Covid19-dataset/train/Covid/01.jpeg")

imgNormal = cv2.imread("Covid19-dataset/train/Normal/01.jpeg")

imgViral = cv2.imread("Covid19-dataset/train/Viral Pneumonia/01.jpeg")
```

### Aplicaci√≥n de Blur

```python
#BLUR for Covid

blurCovid = cv2.blur(imgCovid,(5,5))

plt.subplot(121),plt.imshow(imgCovid),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blurCovid),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#BLUR for Normal

blurNormal = cv2.blur(imgNormal,(5,5))
plt.subplot(121),plt.imshow(imgNormal),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blurNormal),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#BLUR for Viral
blurViral = cv2.blur(imgViral,(5,5))
plt.subplot(121),plt.imshow(imgViral),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blurViral),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

### Aplicaci√≥n de filtro SOBEL

```python
#SOBEL for Covid 

#esta mejora los bordes dependiendo de del axis

sobelxCovid = cv2.Sobel(src=blurCovid,ddepth=cv2.CV_64F, dx=1, dy=0, ksize=3)

sobelyCovid = cv2.Sobel(src=blurCovid,ddepth=cv2.CV_64F, dx=0, dy=1, ksize=3)

sobelCovid = cv2.Sobel(src=blurCovid,ddepth=cv2.CV_64F, dx=1, dy=1, ksize=3)

plt.subplot(221), plt.imshow(imgCovid), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(sobelxCovid), plt.title('Sobel en X')
plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(sobelyCovid), plt.title('Sobel en Y')
plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow(sobelCovid), plt.title('Sobel en XY')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#SOBEL for Normal

#esta mejora los bordes dependiendo de del axis

sobelxNormal = cv2.Sobel(src=blurNormal,ddepth=cv2.CV_64F, dx=1, dy=0, ksize=3)

sobelyNormal = cv2.Sobel(src=blurNormal,ddepth=cv2.CV_64F, dx=0, dy=1, ksize=3)

sobelNormal = cv2.Sobel(src=blurNormal,ddepth=cv2.CV_64F, dx=1, dy=1, ksize=3)

plt.subplot(221), plt.imshow(imgNormal), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(sobelxNormal), plt.title('Sobel en X')
plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(sobelyNormal), plt.title('Sobel en Y')
plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow(sobelNormal), plt.title('Sobel en XY')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#SOBEL for Viral

#esta mejora los bordes dependiendo de del axis


sobelxViral = cv2.Sobel(src=blurViral,ddepth=cv2.CV_64F, dx=1, dy=0, ksize=3)

sobelyViral = cv2.Sobel(src=blurViral,ddepth=cv2.CV_64F, dx=0, dy=1, ksize=3)

sobelViral = cv2.Sobel(src=blurViral,ddepth=cv2.CV_64F, dx=1, dy=1, ksize=3)

plt.subplot(221), plt.imshow(imgViral), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(sobelxViral), plt.title('Sobel en X')
plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(sobelyViral), plt.title('Sobel en Y')
plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow(sobelViral), plt.title('Sobel en XY')
plt.xticks([]), plt.yticks([])
plt.show()

 #is used to change the pixel intensity value of an image
 # based on the surrounding pixel intensity values. This function is used to enhance or remove certain features of an image and create new images
```

### Aplicaci√≥n de filtro Canny

```python
#Canny for Covid

blurG = cv2.GaussianBlur(imgCovid, (3,3), 0,0)
cannyCovid = cv2.Canny(image=blurG, threshold1=35, threshold2=70) 


plt.subplot(221), plt.imshow(imgCovid), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(cannyCovid), plt.title('Canny')
plt.xticks([]), plt.yticks([])
plt.show()

```

```python
#Canny for Normal

blurG = cv2.GaussianBlur(imgNormal, (3,3), 0,0)
cannyNormal = cv2.Canny(image = blurG, threshold1=35, threshold2=70)

plt.subplot(221), plt.imshow(imgNormal), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(cannyNormal), plt.title('Canny')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Canny for Viral

blurG = cv2.GaussianBlur(imgViral, (3,3), 0,0)
cannyViral = cv2.Canny(image = blurG, threshold1=35, threshold2=70)

plt.subplot(221), plt.imshow(imgViral), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(cannyViral, cmap= 'gray'), plt.title('Canny')
plt.xticks([]), plt.yticks([])
plt.show()
```

### Aplicaci√≥n del Filtro seleccionado

```python
#2d for Covid
kernel = np.ones((6,6),np.float32)/25  


filtro2DCovid = cv2.filter2D(imgCovid,-1,kernel)                    

plt.subplot(221), plt.imshow(imgCovid), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(filtro2DCovid), plt.title('2D filter')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#2d for Normal
kernel = np.ones((6,6),np.float32)/25  


filtro2DNormal = cv2.filter2D(imgNormal,-1,kernel)                    

plt.subplot(221), plt.imshow(imgNormal), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(filtro2DNormal), plt.title('2D filter')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#2d for Viral
kernel = np.ones((6,6),np.float32)/25  


filtro2DViral = cv2.filter2D(imgViral,-1,kernel)                    

plt.subplot(221), plt.imshow(imgViral), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(filtro2DViral), plt.title('2D filter')
plt.xticks([]), plt.yticks([])
plt.show()
```
